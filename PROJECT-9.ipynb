```python
#Again checking for missing values
df.isnull().sum()
```

There are two missing values present in 'Lat' and 'Long' and since it won't have any impact on target variable so we are dropping them

```python
#Dropping the column 'Lat','Long_'
df.drop(['Lat','Long_'],axis=1,inplace=True)
df
```

# Model Training

```python
#Splitting data into input and output variable
x=df.drop(['Deaths'],axis=1)
x.head()
```

```python
#Importing all the libraries,metrices required for ML 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=42) #splitting the data into training and testing
```

```python
#checking the shape of training and testing data
print(x_train.shape,x_test.shape)
```

```python
print(y_train.shape,y_test.shape)
```

```python
from sklearn.metrics import r2_score
#Defining a function maxr2_score and will be used for testing different models
def maxr2_score(regr,x,y): 
    max_r_score=0
    for r_state in range(1,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.33)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("Max r2 score corresponding to ",final_r_state,"is",max_r_score)
    return final_r_state
```

```python
#USING THE MODEL LINEAR REGRESSION and calling the function to check r2_score
lreg=LinearRegression()
r_state=maxr2_score(lreg,x,y)
```

```python
#checking the score for the random state 47 
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=47,test_size=0.33)
lreg=LinearRegression()
lreg.fit(x_train,y_train)
lreg.score(x_train,y_train)
```

```python
#checking cross_val_score for LinearRegression
from sklearn.model_selection import cross_val_score
print("Mean r2_score for Linear Regression:",cross_val_score(lreg,x,y,cv=10,scoring="r2").mean())
print("Standard deviation in r2_score for Linear Regression:",cross_val_score(lreg,x,y,cv=10,scoring="r2").std())
```

So Linear Regression is working well and its neither underfitting/overfitting so we can choose it as our final model.

```python
#predict the value
pred=lreg.predict(x_test)
print("Predicted result price:",pred)
print("actual price:",y_test)
```

```python
#checking other metrices
print("error:")

print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:', mean_squared_error(y_test,pred))

print('Root Mean squared error:',np.sqrt(mean_squared_error(y_test,pred)))
```
